```python
import numpy as np
import pandas as pd
```

```python
f1=pd.read_excel("sample_data/Worksheet_student.xlsx", sheet_name="Marks")
f1.head()
```

```python
f2=pd.read_excel("sample_data/Worksheet_student.xlsx", sheet_name="Attendance")
f2.head()
```

```python


cols = ["Mini Test 1", "Mini Test 2", "Live Test", "Assignment"]

f1[cols] = f1[cols].apply(pd.to_numeric, errors="coerce")
f1[cols] = f1[cols].fillna(0)

f1["Total"] = f1[cols].sum(axis=1)

f1

```

```python
f1["%Marks"]=((f1["Total"]/55.0)*100).round(1)
f1
```

```python
df = pd.merge(f1, f2, on="Name")
df
```

```python
attendance_cols = [col for col in df.columns if 'Attendance Day' in str(col)]
df[attendance_cols] = df[attendance_cols].replace({'Y': 1, 'N': 0})
df.head()
```

```python
from numpy._core.defchararray import capitalize
df["Name"]=df["Name"].str.title()
df.head()
```

```python
attendance_cols = [col for col in df.columns if 'Attendance Day' in str(col)]
df['Total Attendance Days'] = df[attendance_cols].sum(axis=1)
df['Attendance Percentage'] = (df['Total Attendance Days'] / len(attendance_cols) * 100).round(1)
df.head()
```

```python
# Define maximum scores for each component to convert them to percentages
max_mini_test_1 = 10
max_mini_test_2 = 10
max_live_test = 15
max_assignment = 20

# Calculate percentage for each component
df['Mini Test 1 %'] = (df['Mini Test 1'] / max_mini_test_1) * 100
df['Mini Test 2 %'] = (df['Mini Test 2'] / max_mini_test_2) * 100
df['Live Test %'] = (df['Live Test'] / max_live_test) * 100
df['Assignment %'] = (df['Assignment'] / max_assignment) * 100

# Calculate the weighted percentage
df['Weighted Percentage'] = (
    (df['Attendance Percentage'] * 0.40) +
    (df['Mini Test 1 %'] * 0.10) +
    (df['Mini Test 2 %'] * 0.10) +
    (df['Live Test %'] * 0.20) +
    (df['Assignment %'] * 0.20)
).round(1)

# Display the relevant columns for verification
df[['Name', 'Attendance Percentage', 'Mini Test 1 %', 'Mini Test 2 %', 'Live Test %', 'Assignment %', 'Weighted Percentage']]
```
